```python
import pandas as pd
from mqtt_write_sql import MessageSolver

%load_ext autoreload
%autoreload 2
```

```python
ms = MessageSolver()
df = ms.get_log('lanzhou')
df
```

```python
df['dtime'] = df['total_sec'].diff()
df['dtime'] = df.apply(lambda row: 0 if row['dtime'] > -300 else 1, axis=1)
df['dtime'] = df['dtime'].cumsum()
df
```

```python
dfg = df.groupby('dtime')
dfs = [dfg.get_group(label) for label in dfg.groups.keys()]

for i in dfs:
    if len(i) != 4:
        print(i.to_string())
df
```
